```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset("tips")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Distribution of tips
sns.histplot(df['tip'], kde=True)
plt.title('Distribution of Tips')
plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Tip amount by day
sns.boxplot(x='day', y='tip', data=df)
plt.title('Tip Amount by Day')
plt.xlabel('Day')
plt.ylabel('Tip Amount')
plt.show()
```

```python
# Tip amount by time of day
sns.boxplot(x='time', y='tip', data=df)
plt.title('Tip Amount by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Tip Amount')
plt.show()
```

```python
# Relationship between total bill and tip
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title('Total Bill vs. Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip Amount')
plt.show()
```

```python
# Tips by size of the party
sns.boxplot(x='size', y='tip', data=df)
plt.title('Tips by Size of the Party')
plt.xlabel('Size of the Party')
plt.ylabel('Tip Amount')
plt.show()
```

```python
# Distribution of total bill
sns.histplot(df['total_bill'], kde=True)
plt.title('Distribution of Total Bills')
plt.xlabel('Total Bill Amount')
plt.ylabel('Frequency')
plt.show()

```

```python
# Swarm plot of tips by gender
sns.swarmplot(x='sex', y='tip', data=df)
plt.title('Tip Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Tip Amount')
plt.show()
```

```python
# Box plot of tips by smoking status
sns.boxplot(x='smoker', y='tip', data=df)
plt.title('Tip Amount by Smoking Status')
plt.xlabel('Smoker')
plt.ylabel('Tip Amount')
plt.show()
```

```python
df['tip_percentage'] = (df['tip'] / df['total_bill']) * 100

sns.boxplot(x='sex', y='tip_percentage', data=df)
plt.title('Tip Percentage by Gender')
plt.xlabel('Gender')
plt.ylabel('Tip Percentage')
plt.show()
```
